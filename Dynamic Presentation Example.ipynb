```r
data(mtcars)
mtcars$cyl = factor(mtcars$cyl)
```

### Data Summary

```r
summary(mtcars)
```

### Case Study

```r
data = mtcars$mpg
variableName = "MPG"
hist(data, main = variableName, col = "green")
```

```r
plot(density(data), main = variableName, col = "red", lwd = 2)
```

```r
par(mfrow = c(2,2))
temp = sapply(levels(mtcars$cyl), function(cyl_level){
    plot(density(data[cyl_level == mtcars$cyl]), main = paste(variableName, "for", cyl_level, "Cylinders"), col = cyl_level)
})
```
